```python
import random

wordlist = []
with open ('wordlist.txt', 'rt') as file:
    for line in file:
         wordlist.append(line.rstrip('\n'))

max_length = int(input("Enter max length of the word you'll have to guess :"))

def choose_word():
    words = [word for word in wordlist if len(word) <= max_length]
    return random.choice(words) if words else None

random_word = choose_word()
word_len = len(random_word)

word = []
for letter in random_word:
    word.append(letter)
    
blank = ["_"] * word_len
print(blank)

lives = 7

HANGMAN_PICS = ['''
    +---+
    O   |
   /|\  |
   / \  |
       ===''','''
    +---+
    O   |
   /|\  |
   /    |
       ===''','''
    +---+
    O   |
   /|\  |
        |
       ===''','''
    +---+
    O   |
   /|   |
        |
       ===''','''
    +---+
    O   |
    |   |
        |
       ===''','''
    +---+
    O   |
        |
        |
       ===''','''
    +---+
        |
        |
        |
       ===''']

game_over = False

while not game_over:
    
    answer = input("enter a letter :").lower()

    for i in range(word_len):
        if answer == word[i]:
            blank[i] = word[i]
            print(blank)
            
    if answer not in word:
        lives -= 1
        print(HANGMAN_PICS[lives])
        
    
    if "_" not in blank:
        game_over = True
        print("you Won")
    
    if lives == 0:
        print("you Lose")
        break
```
